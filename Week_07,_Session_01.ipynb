```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(img, title="image"):
    plt.figure(figsize=(5,5))
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
# Create black image
img_square = np.zeros((200, 200), dtype=np.uint8)

# Create white square
img_square[60:140, 60:140] = 255

show(img_square, "Original square")
```

```python
kernel_x = np.array([[-1, 1]], dtype=np.float32)

grad_square = cv2.filter2D(img_square, cv2.CV_64F, kernel_x)

grad_square_abs = np.abs(grad_square).astype(np.uint8)

show(grad_square_abs, "Gradient edges")
```

```python
# Move square to new position
img_square_moved = np.zeros((200, 200), dtype=np.uint8)
img_square_moved[30:110, 100:180] = 255

show(img_square_moved, "Moved square")
```

```python
grad_moved = cv2.filter2D(img_square_moved, cv2.CV_64F, kernel_x)

grad_moved_abs = np.abs(grad_moved).astype(np.uint8)

show(grad_moved_abs, "Gradient after moving square")
```

```python
img_square_dark = np.zeros((200, 200), dtype=np.uint8)
img_square_dark[60:140, 60:140] = 120

show(img_square_dark, "Darker square")
```

```python
grad_dark = cv2.filter2D(img_square_dark, cv2.CV_64F, kernel_x)

grad_dark_abs = np.abs(grad_dark).astype(np.uint8)

show(grad_dark_abs, "Gradient of darker square")
```

```python
img_square = np.zeros((200,200), dtype=np.uint8)
img_square[60:140,60:140] = 255

show(img_square, "Square image")
```

```python
img_circle = np.zeros((200,200), dtype=np.uint8)

cv2.circle(img_circle, (100,100), 40, 255, -1)

show(img_circle, "Circle image")
```

```python
sobelx_square = cv2.Sobel(img_square, cv2.CV_64F, 1, 0, ksize=3)
sobely_square = cv2.Sobel(img_square, cv2.CV_64F, 0, 1, ksize=3)

show(np.abs(sobelx_square), "Sobel X - Square")
show(np.abs(sobely_square), "Sobel Y - Square")
```

```python
sobelx_circle = cv2.Sobel(img_circle, cv2.CV_64F, 1, 0, ksize=3)
sobely_circle = cv2.Sobel(img_circle, cv2.CV_64F, 0, 1, ksize=3)

show(np.abs(sobelx_circle), "Sobel X - Circle")
show(np.abs(sobely_circle), "Sobel Y - Circle")
```

```python
img_step = np.zeros((200,200), dtype=np.uint8)

img_step[:, :100] = 50
img_step[:, 100:] = 200

show(img_step, "Step image")
```

```python
sobelx = cv2.Sobel(img_step, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(img_step, cv2.CV_64F, 0, 1, ksize=3)

sobel_mag = np.sqrt(sobelx**2 + sobely**2)

show(sobel_mag, "Sobel magnitude")
```

```python
laplacian1 = cv2.Laplacian(img_step, cv2.CV_64F, ksize=1)

show(np.abs(laplacian1), "Laplacian ksize=1")
```

```python
laplacian3 = cv2.Laplacian(img_step, cv2.CV_64F, ksize=3)

show(np.abs(laplacian3), "Laplacian ksize=3")
```

```python
show(sobel_mag, "Sobel magnitude")
show(np.abs(laplacian3), "Laplacian ksize=3")
```

```python
img_step = np.zeros((200,200), dtype=np.uint8)

img_step[:, :100] = 50
img_step[:, 100:] = 200

show(img_step, "Step image")
```

```python
edges1 = cv2.Canny(img_step, 30, 100)

show(edges1, "Canny 30-100")
```

```python
edges2 = cv2.Canny(img_step, 50, 150)

show(edges2, "Canny 50-150")
```

```python
edges3 = cv2.Canny(img_step, 80, 200)

show(edges3, "Canny 80-200")
```

```python
edges4 = cv2.Canny(img_step, 100, 250)

show(edges4, "Canny 100-250")
```

```python
#The results look similar because the step image contains a very strong edge. All threshold pairs detect the same boundary.
```

```python
# create noisy step image
img_noise = img_step.copy().astype(np.float32)

noise = np.random.normal(0, 20, img_noise.shape)

img_noise = img_noise + noise
img_noise = np.clip(img_noise, 0, 255).astype(np.uint8)

show(img_noise, "Noisy step image")
```

```python
edges_low = cv2.Canny(img_noise, 20, 60)

show(edges_low, "Canny 20-60")
```

```python
edges_mid = cv2.Canny(img_noise, 50, 150)

show(edges_mid, "Canny 50-150")
```

```python
edges_high = cv2.Canny(img_noise, 120, 250)

show(edges_high, "Canny 120-250")
```
